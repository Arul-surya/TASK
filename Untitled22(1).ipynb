```python
df8_Table1
```

```python
df9_Table1 = df8_Table1.groupby(['Campaign','Metric'])['Value'].sum().reset_index(name='Total Impressions')
```

```python
df9_Table1
```

```python
df10_Table1 = df9_Table1.sort_values(by='Total Impressions', ascending=False)
```

```python
df10_Table1
```

```python
df11_Table1
```

```python
df12_Table1 = df11_Table1.groupby(['Campaign','Metric'])['Value'].sum().reset_index(name='Total Media Cost')
df12_Table1
```
